```python
import pandas as pd
import soynlp
import string
```

```python
# 데이터 읽기

file_path = 'naver_titles_0701_0710.csv'
corpus_path = 'naver_title_dic.csv'

df = pd.read_csv(file_path, encoding='utf-8', header=None, names=['Title'])
df['Title'] = df['Title'].dropna().apply(lambda x:''.join([i for i in x 
                                                  if i not in string.punctuation]))
df.to_csv(corpus_path, encoding='utf-8', index=False)
```

```python
from soynlp.utils import DoublespaceLineCorpus
from soynlp.noun import LRNounExtractor_v2

sents = DoublespaceLineCorpus(corpus_path, iter_sent=True)
```

```python
noun_extractor = LRNounExtractor_v2(verbose=True, extract_compound=True)
noun_extractor.train(sents)
nouns = noun_extractor.train_extract(sents, min_noun_frequency=1,
                                     min_noun_score=0.3)
```

```python
a = [noun for noun, score in nouns.items() if len(noun) > 1]
a = [i.replace('"', '') for i in a] 
a = [i.replace("'", '') for i in a] 
len(a)
```

```python
a = pd.DataFrame(a, columns=['word'])
a['pos'] = 'NNP'
a = a[a['word'].str[-1] != '들']
a = a.drop_duplicates()
a
```

```python
a.to_csv(corpus_path, encoding='utf-8', index=False, header=False)
```
