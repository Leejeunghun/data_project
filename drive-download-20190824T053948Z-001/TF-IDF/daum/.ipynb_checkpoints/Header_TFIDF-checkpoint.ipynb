```python
import pandas as pd
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from future.utils import iteritems
from collections import Counter
from sklearn.manifold import TSNE
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
article_data = pd.read_csv('selected_Title_date_from15to17.csv', encoding='utf-8',
                          header=None)
documents = [' '.join(i[0].split(' ')[1:]) for i in article_data.values] 
```

```python
as_one = ''
for document in documents:
    as_one = as_one + ' ' +document
words = as_one.split()
```

```python
words[0:10]
```

```python
counts = Counter(words)
counts
```

```python
vocab = sorted(counts, key=counts.get, reverse = True)
vocab
```

```python
word2idx = {word.encode("utf8").decode("utf8"): ii for ii, word in enumerate(vocab,1)}
word2idx
```

```python
idx2word = {ii: word for ii, word in enumerate(vocab)}
idx2word
```

```python
V = len(word2idx)
N = len(documents)

tf = CountVectorizer()
tf.fit_transform(documents)
tf.fit_transform(documents)[0:1].toarray()
```

```python
tfidf = TfidfVectorizer(max_features=100, max_df=0.95, min_df=0.1)
A_tfidf_sp = tfidf.fit_transform(documents)

tfidf_dict = tfidf.get_feature_names()
print(tfidf_dict)
```

```python
data_array = A_tfidf_sp.toarray()
data = pd.DataFrame(data_array, columns=tfidf_dict)
data.shape
```

```python
tsne = TSNE(n_components=2, n_iter=10000, verbose=1)
data_array.shape
```

```python
data_array.T.shape
```

```python
Z = tsne.fit_transform(data_array.T)
```

```python
print(Z[0:5])
print('Top Words: ', len(Z))
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib as mpl
mpl.rcParams.update({'font.size': 20})

plt.figure(figsize=(38,28))
size = tfidf.idf_ * 200
plt.scatter(Z[:,0], Z[:,1], s=size)
for x, y, name in zip(Z[:,0], Z[:,1], tfidf_dict):
    plt.text(x, y, name)
plt.show()
```
